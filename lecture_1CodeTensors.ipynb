```python
N = 16
x = np.ones((N, N))
w = np.ones((N, N))
```

```python
def slow_matmul() -> np.ndarray:
    y = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            for k in range(N):
                y[i, j] += x[i, k] * w[k, j]
    return y
```

```python
python_time = timeit.timeit(slow_matmul, number=1)
print(f"Python matmul took {python_time:.4f} seconds")
```

```python
numpy_time = timeit.timeit(lambda: x @ w, number=1)
print(f"Numpy matmul took {numpy_time:.4f} seconds")
```

```python
speedup = python_time / numpy_time
print(speedup)
```
