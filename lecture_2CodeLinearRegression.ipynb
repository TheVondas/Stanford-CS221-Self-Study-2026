```python
@dataclass
class Example1D:
    input: float
    output: float
```

```python
def get_training_data():
    return [
    Example1D(input=1.0, output=4.0),
    Example1D(input=2.0, output=6.0),
    Example1D(input=4.0, output=7.0),
    ]
```

```python
training_data = get_training_data()
```

```python
@dataclass
class Parameters1D:
    weight: float
    bias: float
```

```python
def fixed_f(x: float) -> float:
    return 2 * x + 1
```

```python
params = Parameters1D(weight=3.0, bias=1.0)
```

```python
def f(params: Parameters1D, x: float) -> float:
    return params.weight * x + params.bias
```

```python
x1 = 1
y1 = f(params, x1)
print(y1)
```

```python
params = Parameters1D(weight=2.0, bias=0.2)
```

```python
x1 = 1
y1 = f(params, x1)
print(y1)
```

```python
def compute_loss(params: Parameters1D, example: Example1D) -> float:
    residual = f(params, example.input) - example.output
    return residual ** 2
```

```python
def compute_train_loss(params: Parameters1D, training_data: list[Example1D]) -> float:
    losses = [compute_loss(params, example) for example in training_data]
    train_loss = np.mean(losses)
    return train_loss
```

```python
def compute_grad_loss(params: Parameters1D, example: Example1D) -> np.ndarray:
    residual = (params.weight * example.input + params.bias) - example.output
    loss = residual ** 2
    grad_weight = 2 * residual * example.input
    grad_bias = 2 * residual
    return np.array([grad_weight, grad_bias])
```

```python
def compute_gradient_train_loss(params: Parameters1D, training_data: list[Example1D]) -> np.ndarray:
    grads = [compute_grad_loss(params, example) for example in training_data]
    grad = np.mean(grads, axis=0)
    return grad
```

```python
params = Parameters1D(weight=2.0, bias=1.0)
values = [{"x": x, "y": f(params, x)} for x in np.linspace(0, 5, 30)]

Chart(Data(values = values)).mark_line().encode(x="x:Q", y="y:Q")
```

```python
params = Parameters1D(weight=0, bias=1.0)
training_data = get_training_data()
train_loss = compute_train_loss(params, training_data)
print(train_loss)
```

```python
grad = compute_gradient_train_loss(params, training_data)
print(grad)
```

```python
learning_rate = 0.01

params = Parameters1D(
weight = params.weight - learning_rate * grad[0],
bias = params.bias - learning_rate * grad[1])
```

```python
new_train_loss = compute_train_loss(params, training_data)
print(new_train_loss)
```
